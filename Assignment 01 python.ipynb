3. Write a program in which use all the operators we can use in Python

```python
a = 20
b = 6

# Arithmetic Operators
print("=== Arithmetic Operators ===")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)

# Comparison Operators
print("\n=== Comparison Operators ===")
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)

# Logical Operators
print("\n=== Logical Operators ===")
print("a > 5 and b > 2:", a > 5 and b > 2)
print("a > 5 or b > 10:", a > 5 or b > 10)
print("not(a > 5):", not(a > 5))

# Assignment Operators
print("\n=== Assignment Operators ===")
x = 10
x += 5;  print("+=:", x)
x -= 3;  print("-=:", x)
x *= 2;  print("*=:", x)
x //= 4; print("//=:", x)

# Bitwise Operators
print("\n=== Bitwise Operators ===")
print("a & b:", a & b)
print("a | b:", a | b)
print("a ^ b:", a ^ b)
print("~a:", ~a)
print("a << 1:", a << 1)
print("a >> 1:", a >> 1)

# Identity Operators
print("\n=== Identity Operators ===")
p = [1, 2]
q = p
print("p is q:", p is q)
print("p is not q:", p is not [1, 2])

# Membership Operators
print("\n=== Membership Operators ===")
my_list = [1, 2, 3, 4, 5]
print("3 in my_list:", 3 in my_list)
print("10 not in my_list:", 10 not in my_list)
```

4. Completes the following steps of small task:
    - Mention Marks of English , Islamiat and Maths out of 100 in 3 different variables
    - Mention Variable of Total Marks and assign 300 to it
    - Calculate Percentage

```python
english = 85
islamiat = 90
maths = 78

total_marks = 300
obtained_marks = english + islamiat + maths

percentage = (obtained_marks / total_marks) * 100

print("English Marks:", english)
print("Islamiat Marks:", islamiat)
print("Maths Marks:", maths)
print("Total Obtained:", obtained_marks, "/ 300")
print("Percentage:", percentage, "%")
```
